```python
dbutils.widgets.text('filename', 'bookings')
filename = dbutils.widgets.get('filename')
```

```python
filename
```

```python
df = spark.readStream.format('cloudFiles')\
                .option('cloudFiles.format', 'csv')\
                .option('cloudFiles.schemaLocation', f'/Volumes/dbx_dbt_flight/bronze/bronze_volume/{filename}/checkpoint')\
                .option('cloudFiles.schemaEvolutionMode', 'rescue')\
                .load(f'/Volumes/dbx_dbt_flight/raw/raw_data/{filename}/')
```

```python
df.writeStream.format('delta')\
            .outputMode('append')\
            .option('checkpointLocation', f'/Volumes/dbx_dbt_flight/bronze/bronze_volume/{filename}/checkpoint')\
            .option('path', f'/Volumes/dbx_dbt_flight/bronze/bronze_volume/{filename}/data')\
            .trigger(once=True)\
            .start()
    
```
